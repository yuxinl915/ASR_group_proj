```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace 'your_file.csv' with the path to your CSV file
file_path = 'dataset/s1_r_1/s1_r_1-paper-0-emg.csv'

# Replace 'Column_Name' with the name of the column you want to plot
column_to_plot = '6'

# Read the CSV file
data = pd.read_csv(file_path)

# Plot the column
plt.figure(figsize=(10, 6))
plt.plot(data[column_to_plot], marker='o')
plt.title(f'Plot of {column_to_plot}')
plt.xlabel('Index')
plt.ylabel(column_to_plot)
plt.grid(True)
plt.show()
```
